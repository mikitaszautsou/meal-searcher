```python
# Extract all nutrient names from all foods in format like this:
# nutrients = {
#     'Oats': {'Calories': 389, 'Protein': 16.9, 'Fat': 6.9, 'Carbs': 66.3},
#     'Milk': {'Calories': 42, 'Protein': 3.4, 'Fat': 1.0, 'Carbs': 5.0},
#     'Banana': {'Calories': 89, 'Protein': 1.1, 'Fat': 0.3, 'Carbs': 22.8}
# }
# data['SurveyFoods'][0]['foodNutrients'][0]
nutrients = {}
for food in data['SurveyFoods']:
    food_name = food['description']
    nutrients[food_name] = {}
    for nutrient in food['foodNutrients']:
        nutrient_name = nutrient['nutrient']['name']
        nutrient_value = nutrient['amount']
        nutrients[food_name][nutrient_name] = nutrient_value



# define function that build dictionary with food name as key and dictionary of nutrients as value, but keep only provided nutrient names
def build_nutrients(data, nutrient_names):
    nutrients = {}
    measurements = {}
    for food in data['SurveyFoods']:
        food_name = food['description']
        nutrients[food_name] = {}
        for nutrient in food['foodNutrients']:
            nutrient_name = nutrient['nutrient']['name']
            if nutrient_name in nutrient_names:
                nutrient_value = nutrient['amount']
                nutrients[food_name][nutrient_name] = nutrient_value
                if nutrient_name not in measurements:
                    measurements[nutrient_name] = nutrient['nutrient']['unitName']
                elif measurements[nutrient_name] != nutrient['nutrient']['unitName']:
                    raise ValueError(f'Inconsistent unit for {nutrient_name}: {measurements[nutrient_name]} and {nutrient["nutrient"]["unitName"]}')
    return nutrients, measurements

nutreints, measurements = build_nutrients(data, ['Energy'])
measurements
```

```python
import pulp as pl

# Define the problem
prob = pl.LpProblem("Meal_Planning_Problem", pl.LpMinimize)

# Create a dictionary of pulp variables with lower limit
food_vars = pl.LpVariable.dicts("Amount", foods, lowBound=0)

# Objective Function: minimize the total weight of foods
prob += pl.lpSum([food_vars[f] for f in foods]), "Total Weight of Food"

def add_nutrient_constraints(prob, nutrients, food_vars, nutrient_name, min_value, max_value):
    if min_value is not None:
        prob += pl.lpSum([nutrients[f][nutrient_name] * food_vars[f] for f in foods]) >= min_value, f"Min {nutrient_name}"
    if max_value is not None:
        prob += pl.lpSum([nutrients[f][nutrient_name] * food_vars[f] for f in foods]) <= max_value, f"Max {nutrient_name}"
calories = 2500
calories = 1600
add_nutrient_constraints(prob, nutrients, food_vars, 'Energy', calories, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Carbohydrate, by difference', calories * 0.45 / 4,  calories * 0.65 / 4)
add_nutrient_constraints(prob, nutrients, food_vars, 'Protein', calories * 0.1 / 4,  calories * 0.35 / 4)
add_nutrient_constraints(prob, nutrients, food_vars, 'Total lipid (fat)', calories * 0.0 / 9,  calories * 0.3 / 9)
add_nutrient_constraints(prob, nutrients, food_vars, 'Fiber, total dietary', 25, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Sugars, total including NLEA', None, 36)
add_nutrient_constraints(prob, nutrients, food_vars, 'Sodium, Na', None, 1500)
add_nutrient_constraints(prob, nutrients, food_vars, 'Cholesterol', None, 300)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin A, RAE', 900, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin C, total ascorbic acid', 90, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin D (D2 + D3)', 20, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin E (alpha-tocopherol)', 15, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin K (phylloquinone)', 120, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Folate, DFE', 400, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin B-6', 1.3, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Vitamin B-12', 2.4, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Thiamin', 1.2, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Riboflavin', 1.3, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Niacin', 16, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Calcium, Ca', 1200, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Iron, Fe', 8, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Magnesium, Mg', 420, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Zinc, Zn', 11, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Potassium, K', 3400, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Fatty acids, total saturated', None, 20)
add_nutrient_constraints(prob, nutrients, food_vars, 'Selenium, Se', 55, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Carotene, beta', 10800, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Carotene, alpha', 21100, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Lycopene', 15000, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Copper, Cu', 1.4, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Phosphorus, P', 700 , None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Choline, total', 550, None)
add_nutrient_constraints(prob, nutrients, food_vars, 'Lutein + zeaxanthin', 10000, None)# can be lowered to 6000
add_nutrient_constraints(prob, nutrients, food_vars, 'Folic acid', 400, None)


# Solve the problem
prob.solve()
# Status of the problem
print("Status:", pl.LpStatus[prob.status])
if pl.LpStatus[prob.status] != 'Optimal':
    raise ValueError("Could not find optimal solution")
# Output the results for each food variable with a non-zero value, formatted nicely
for v in prob.variables():
    if v.varValue > 0:
        # Remove 'Amount_' prefix and replace underscores with spaces in the food name
        formatted_name = v.name.replace("Amount_", "").replace("_", " ")
        print(f"{formatted_name} = {v.varValue * 100} grams")

# Print a divider line for clarity
print('-' * 40)

# Display the calculated nutrient totals
for nutrient_name in measurements:
    nutrient_sum = sum(nutrients[f][nutrient_name] * food_vars[f].varValue for f in foods)
    print(f"{nutrient_name}: {nutrient_sum} {measurements[nutrient_name]}")

# TODO: deversify products
```
