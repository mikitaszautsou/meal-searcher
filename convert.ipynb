```python
# load SR data

import json

with open('data/FoodData_Central_sr_legacy_food_json_2021-10-28.json') as f:
    sr_data = json.load(f)
```

```python
#  load foundation data
with open('data/foundationDownload.json') as f:
    foundation_data = json.load(f)
```

```python
import copy
# foundation_data['FoundationFoods'][0]['ndbNumber']
# sr_data['SRLegacyFoods'][0]['ndbNumber']

# iterate over all fondation data and find corresponding sr data and keep it in the separate array
combined_data = []
not_found_items_count = 0
for foundation_food in foundation_data['FoundationFoods']:
    foundation_food_combined = copy.deepcopy(foundation_food)
    food_found = False
    for sr_food in sr_data['SRLegacyFoods']:
        if foundation_food['ndbNumber'] == sr_food['ndbNumber']:
            food_found = True
            for sr_nutrient in sr_food['foodNutrients']:
                nutrient_found = False
                nutrient_id = sr_nutrient['nutrient']['id']
                for foundation_nutrient in foundation_food['foodNutrients']:
                    if nutrient_id == foundation_nutrient['nutrient']['id']:
                        nutrient_found = True
                        if foundation_nutrient['amount'] != sr_nutrient['amount'] and foundation_nutrient['amount'] <= 0.0001:
                            print(f'Warning: {foundation_food["description"]} {foundation_nutrient["nutrient"]["name"]} {foundation_nutrient["amount"]} != {sr_nutrient["amount"]}')
                if not nutrient_found:
                    foundation_food_combined['foodNutrients'].append(sr_nutrient)
    combined_data.append(foundation_food_combined)
    # if not food_found:
    #     not_found_items_count += 1
    #     print(foundation_food['description'])

print(f'Not found items count: {not_found_items_count / len(foundation_data["FoundationFoods"])}')
```

```python
for food in combined_data:
    if food['description'] == 'Strawberries, raw':
        print(food)
        break
```

```python
# save combined_data to json
with open('data/combined_data.json', 'w') as f:
    json.dump(combined_data, f, indent=2)
```
